```python
#Answer-5

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(42)
n_samples = 100

# Create correlated features
x1 = np.random.rand(n_samples)  # First feature
x2 = x1 * 0.7 + np.random.rand(n_samples) * 0.3  # Second feature correlated with the first
x3 = x1 * 0.3 + np.random.rand(n_samples) * 0.7  # Third feature with some correlation
x4 = np.random.rand(n_samples)  # Fourth feature independent of the others

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()


#A synthetic dataset with 100 samples and four features is created, where Feature 2 and Feature 3 are correlated with Feature 1.
#A correlation matrix is computed for the dataset.
#A heatmap is generated to visualize the correlation matrix, with annotations showing the correlation coefficients.
```

```python
#Answer-4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the meshgrid and calculate z
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(df['X'].values.reshape(100, 100),
                 df['Y'].values.reshape(100, 100),
                 df['Z'].values.reshape(100, 100),
                 cmap='viridis', edgecolor='none')

# Add title and labels
ax.set_title("3D Surface Plot of z = sin(sqrt(x^2 + y^2))")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show the plot
plt.show()
```

```python
#Answer-5

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),  # Population in millions
    'GDP': np.random.randint(500, 2000, 5)          # GDP in billions
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Bubble size (scaled for better visibility)
            alpha=0.5,
            c='blue',
            edgecolors='w')

# Add title and labels
plt.title("Bubble Chart of Population vs GDP")
plt.xlabel("GDP (in billions)")
plt.ylabel("Population (in millions)")

# Add annotations for each bubble
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 fontsize=9,
                 ha='right')

# Show the plot
plt.grid()
plt.xscale('log')  # Optional: logarithmic scale for better visualization
plt.show()
```

```python
#Answer-2

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable output in Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x values from 0 to 10
y = np.random.rand(n) * 10  # Random y values from 0 to 10
sizes = np.random.randint(10, 100, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors for markers

# Create a DataFrame
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a Bokeh scatter plot
p = figure(title="Random Scatter Plot", width=800, height=400)

# Add a scatter renderer with varying sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=source, alpha=0.6)

# Add title and labels
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.legend.title = "Colors"

# Show the plot
show(p)
```

```python
#Answer-4

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable output in Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource for the histogram
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Create a Bokeh histogram
# Changed plot_height to height and plot_width to width
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400, width=600)

# Add bars to the histogram
p.quad(top='top', bottom=0, left='left', right='right', source=source,
        fill_color='blue', line_color='white')

# Show the plot
show(p)
```

```python
#Answer-5

#Answer-5

!pip install bokeh  # Make sure Bokeh is installed

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256  # Changed from bokeh.palette to bokeh.palettes

# Enable output in Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare the data for Bokeh
data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}
source = ColumnDataSource(data)

# Create a Bokeh heatmap
# Changed plot_height to height and plot_width to width
p = figure(title="Heatmap Example",
           x_axis_label='X',
           y_axis_label='Y',
           height=400, width=400)

# Define a linear color mapper
# Added low and high values for the 'value' field
mapper = linear_cmap('value', Viridis256, low=min(data_heatmap.flatten()), high=max(data_heatmap.flatten()))

# Add rectangles to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
